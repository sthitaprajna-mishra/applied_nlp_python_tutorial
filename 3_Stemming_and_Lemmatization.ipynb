```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import state_union
```

```python
words = word_tokenize(state_union.raw("1961-Kennedy.txt"))
```

```python
ps = PorterStemmer()
```

```python
for i in words:
    print(ps.stem(i))
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()

words = ["cats", "feet", "bats", "corpora"]

for i in words:
    print(lem.lemmatize(i))
```

# Parts-Of-Speech Tagging 

```python
from nltk import pos_tag
```

```python
words = word_tokenize(state_union.raw("1996-Clinton.txt"))
```

```python
pos_tag(words)
```

# Named-Entity-Recognition

```python
from nltk.corpus import gutenberg

pos_words = pos_tag(word_tokenize(gutenberg.raw("edgeworth-parents.txt")))
```

```python
from nltk import ne_chunk
```

```python
ner_words = ne_chunk(pos_words[:1000])
```

```python
print(ner_words)
```
