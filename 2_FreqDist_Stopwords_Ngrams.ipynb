```python
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg
from nltk.probability import FreqDist
```

```python
sample = gutenberg.raw("carroll-alice.txt")
```

```python
word_tokens = word_tokenize(sample)

freq = FreqDist(word_tokens)
```

```python
freq
```

```python
freq.most_common(5)
```

```python
freq.plot(30)
```

```python
for i in freq.most_common(30):
    if i[0] == 'Alice':
        print(i[1])
```

# Stopwords

```python
from nltk.corpus import stopwords

stop_words = stopwords.words("english")

print(stop_words)
```

```python
from nltk.corpus import genesis

sample = genesis.raw("english-web.txt")
```

```python
word_tokens = word_tokenize(sample)
```

```python
#filtered_words = [i for i in word_tokens if i not in stop_words and len(i) > 3]
```

```python
filtered_words = []

for i in word_tokens:
    if i not in stop_words and len(i) > 3:
        filtered_words.append(i)
```

```python
freq = FreqDist(filtered_words)
```

```python
freq.plot(30)
```

```python
freq.most_common(5)
```

# Bigrams

```python
from nltk import bigrams
from nltk.corpus import webtext
```

```python
words = [i.lower() for i in webtext.words("pirates.txt") if i not in stop_words and len(i) > 3]
bigram = bigrams(words)
```

```python
freq = FreqDist(bigram)
```

```python
freq.most_common(5)
```

# Trigrams

```python
from nltk import ngrams
from nltk.corpus import gutenberg

word = [i.lower() for i in gutenberg.words("shakespeare-caesar.txt") if i not in stop_words and len(i) > 3]
```

```python
ngram = ngrams(word, 4)
```

```python
freq = FreqDist(ngram)
```
